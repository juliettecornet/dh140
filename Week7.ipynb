```python
#importing Midsummer Night's Dream 
import requests 
response = requests.get('https://www.gutenberg.org/files/1514/1514-h/1514-h.htm') 
```

```python
response
```

```python
from bs4 import BeautifulSoup 
```

```python
html_string = response.text
document = BeautifulSoup(html_string, "html.parser")
```

```python
print(document.get_text()) #printing just the play
```

```python
text = document.get_text()
```

```python
#tokenizing the play
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent = sent_tokenize(text)
print(sent)
```

```python
print(word_tokenize(sent[7]))
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
#removing the stopwords from the play 
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
#make a string of all the extra words and names not in stopwords than mean nothing 
extrawords = ['â','thou','shall','hermia', 'lysander','demetrius', 'helena', 'project', 'theseus', 'thy', 'thee', 'come', 'gutenberg', 'gutenberg™', 'puck', 'thisbe', 'quince', 'must', 'pyramus', 'may', 'oberon','’','titania', 'hath', 'enter']
myStopWords = list(punctuation) + stopwords.words('english')+ extrawords
myStopWords
```

```python
wordsNoStop = []
for i in words:
    i = i.lower()  
    if i not in myStopWords:
        wordsNoStop.append(i)
# print(words)
print(wordsNoStop)
```

```python
wordsNoStopWords = [w for w in words if w not in myStopWords]
```

```python
wordsNoStopWords
```

```python
from nltk.probability import FreqDist
from string import punctuation
import numpy as np
import pandas as pd
```

```python
from nltk.stem.lancaster import LancasterStemmer
from nltk.stem.porter import PorterStemmer
```

```python
#stemming the words
wordLancasterStems = [LancasterStemmer().stem(w) for w in wordsNoStop]
wordPorterStems = [PorterStemmer().stem(w) for w in wordsNoStop]
```

```python
print(wordLancasterStems)
print(wordPorterStems)
```

```python
#finding the word frquencies 
freq = FreqDist(wordLancasterStems)
```

```python
freq
```

```python
#making a data frame of word frequencies 
df_items = freq.items()
df_list = list(df_items)
df = pd.DataFrame(df_list)
df = df.sort_values(1, ascending = False)
df
```

```python
#turning column 1 into a list to calculate the mean, meadian and mode of frequencies
wordfreq = df[1].tolist()
print(wordfreq)
```

```python
#calculating the mean
np.mean(wordfreq)
```

```python
#calculating the median
import statistics 
np.median(wordfreq)
```

```python
#calculating the mode
statistics.mode(wordfreq)
```

```python
from scipy import stats

#calculating 10% trimmed mean
stats.trim_mean(wordfreq, 0.1)
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#making histogram of word frequencies 
df.hist()
```

```python
#the mean, median, mode and trimmed median would all be found in that first bin between 0 and 20 
```

```python
#calculating the standard deviation
np.std(wordfreq)
```

```python
#calculating the interquartile range 
np.quantile(wordfreq, 0.25)
```

```python
np.quantile(wordfreq, 0.75)
```

```python
q3, q1 = np.percentile(wordfreq, [75 ,25])
iqr = q3 - q1
```

```python
iqr
```

```python
#standard deviation is bigger than the interquartile range. 
#on the histogram, they would all befound in the same bin as the mean, median, mode. 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
X = target_df
arr = X.values
  
# reshaping series 
reshaped_arr = arr.reshape((-1, 1))
  
# show 
y = df.data['bmi']
reg = LinearRegression().fit(reshaped_arr, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
ytrain = reg.intercept_ + reg.coef_ * X
```

```python
plt.plot(X,y,'ro',X,ytrain,'b-');
```

```python
mean_squared_error(y, ytrain)
```

```python
r2_score(y, ytrain)
```

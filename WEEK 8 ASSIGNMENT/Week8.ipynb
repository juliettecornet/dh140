```python
#making character list
characters = [ 
    'HERMIA',
    'LYSANDER',
    'HELENA',
    'DEMETRIUS',
    'THESEUS',
    'HIPPOLYTA',
    'EGEUS',
    'PHILOSTRATE',
    'NICK BOTTOM',
    'PETER QUINCE',
    'FRANCIS FLUTE',
    'TOM SNOUT',
    'SNUG',
    'ROBIN STARVELING',
    'OBERON',
    'TITANIA',
    'ROBIN GOODFELLOW',
    'FAIRY',
    'PEASEBLOSSOM',
    'COBWEB',
    'MOTE',
    'MUSTARDSEED',
    'BOTTOM',
    'FLUTE',
    'QUINCE',
    'ROBIN',
    'STARVELING'
]
```

```python
charnum = {}

connections = {}
for i in range(len(characters)-1):
    for j in range(i+1,len(characters)):
        connections[(characters[i],characters[j])] = 0

for k in characters:
    charnum[k] = 0

for i in acts:
    for j in i.split('Scene')[1:]:
        scenechars = []
        for k in characters:
            if j.find(k) != -1:
                scenechars.append(k)
                charnum[k] += 1
        for a in range(len(scenechars)-1):
            for b in range(a+1,len(scenechars)):
                connections[(scenechars[a],scenechars[b])] += 1
```

Now I would argue that the "most important" nodes are the main characters of the playthe characters that are in the most scenes. Based on the graph, there doesn't to be just one main character in A Midsummer's Night Dream since there are multiple nodes that are around the same size. Based on my knowledge fo the play, I know this to be true since this play is more of an ensemble. The most important nodes based on the graph are therefore: Demetrius, Lysander, Hermia, Titania, Oberon, Robin and Helena.  

However, let's do some quantitative analysis to know for sure. 
Let's say the important nodes are the nodes with the most edges. Therefore those nodes would be the characters with the most scenes in common with other characters. Let's calcuate the network centrialty to figure this out. 

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
nxgraphitems = []

# for k,v in charnum.items():
#     dashnode = k
#     graphitems.append(dashnode)

for k,v in connections.items():
    if v != 0:
        dashedge = (k[0],k[1])
        nxgraphitems.append(dashedge)
    
nxgraphitems
```

```python
G = nx.Graph()
```

```python
G.add_edges_from(nxgraphitems)
```

```python
G.nodes
```

```python
nx.degree_centrality(G)
```

```python
dc = nx.degree_centrality(G)
for i in sorted(dc, key=dc.get, reverse=True):
    print('{:15s}: {:.3f}'.format(i.title(), dc[i]))
```

Based on this degree of centrality, Titania and Robin are the characters with the most scenes in connection with others, making them the most important nodes with centriality scores of 1.00 each. However, when you look at the other degrees of centriality, there are a lot of the same high numbers. For example, Lysander, Demetrius, Oberon and Bottom all have 0.95, putting them right up there with Titania and Robin. Therefore, the degree of centrality does confirm that this is an ensemble play for the most part. 

Something off-topic that I find intersting is that Hermina and Helena have much lower degrees of centrality (0.7) compared to Lysander and Demetrius (0.95), the lovers in question in the play that are argued to be the main point of conflict. The disparity between the scenes between the men versus the women is interesting to note. 

Let's do some more quantitive analysis and look at the network's density. 

```python
nx.density(G)
```

Based on high network density of 0,7, so it has pretty high number of connections compared to maximum number of connections it could potentially have in the play. Once again, this makes sense considering the ensemble nature of the playzit means that a lot of the characters have the oppoutnity to interact with one another in a scene. 

```python
#identify different communities in the play via the graph
from networkx.algorithms.community.centrality import girvan_newman
communities = girvan_newman(G)
from networkx.algorithms.community import greedy_modularity_communities
```

```python
greedy_modularity_communities(G)
```

```python
communities = greedy_modularity_communities(G)
# Create empty dictionary
modularity_class = {}
#Loop through each community in the network
for community_number, community in enumerate(communities):
    #For each member of the community, add their community number
    for name in community:
        modularity_class[name] = community_number
```

```python
 modularity_class
```

```python
communities = greedy_modularity_communities(G)

node_groups = []
for com in communities:
    node_groups.append(list(com))
 
print(node_groups)
 
color_map = []
for node in G:
    if node in node_groups[0]:
        color_map.append('blue')
    elif node in node_groups[1]:
        color_map.append('yellow')
    else:
        color_map.append('green')
nx.draw(G, node_color=color_map, with_labels=True)
plt.show()
```

```python
graphitems = []

for k,v in charnum.items():
    if k in node_groups[0]:
        modularity = 'blue'
    elif k in node_groups[1]:
        modularity = 'green'
    else:
        modularity = 'red'
    dashnode = {'data': {'id': k,
                         'label': k.title(),
                         'size': str(v),'modularity': modularity}}
    graphitems.append(dashnode)

correction_factor = max(connections.values())
print(correction_factor)
for k,v in connections.items():
    if v != 0:
        dashedge = {'data': {'source': k[0],
                             'target': k[1],
                             'weight': v/correction_factor}}
        graphitems.append(dashedge)
    
graphitems
```

```python
app = JupyterDash(__name__)

app.layout = html.Div([
    cyto.Cytoscape(
        layout={'name': 'cose'},
        elements=graphitems,
        style={'width': '100%', 'height': '750px'},
        stylesheet=[
            {
                'selector': 'node',
                'style': {
                    'content':'data(label)',
                    'text-halign':'center',
                    'text-valign':'center',
                    'width': 'data(size)',
                    'height': 'data(size)',
                    'font-size': 4,
                    'color': 'black',
                    'background-color': 'data(modularity)',
                    'text-outline-color': 'white',
                    'text-outline-width': 0.2,
                    'shape':'circle'
                }
            },
            {
                'selector':'edge',
                'style': {
                    'width':'data(weight)',
                    'line-color': 'blue',
                  }
            },
        ]
    )
])

app.run_server(debug=True)
```

The final graph calculates the connections between the the characters in A Midsummer Night's Dream. The nodes of the graph are the charactersthe bigger the node, the more scenes they are in with other characters in the play. The edges are the connections between the characters. The bigger the edge between each node, the more scenes these two characters were in together. 
The nodes are also color coordinated by their communities. Based on the graph, there are two main communities in A Midsummer Night's Dream, identifyiable by green or blue nodes in the graph.

```python
from bs4 import BeautifulSoup 
```

```python
html_string = response.text
document = BeautifulSoup(html_string, "html.parser")
```

```python
print(document.get_text()) #printing just the play
```

```python
text = document.get_text()
```

```python
text
```

```python
#tokenizing the play
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
sent = sent_tokenize(text)
# sent = sent.lower()
print(sent)
```

```python
print(word_tokenize(sent[7]))
```

```python
words = []
for s in sent:
    for w in word_tokenize(s):
        words.append(w)
```

```python
print(words)
```

```python
#removing the stopwords from the play 
from nltk.corpus import stopwords
from string import punctuation
nltk.download('stopwords')
```

```python
#make a string of all the extra words and names not in stopwords than mean nothing 
extrawords = ['â','thou','shall','hermia', 'lysander','demetrius', 'helena', 'project', 'theseus', 'thy', 'thee', 'come', 'gutenberg', 'gutenberg™', 'puck', 'thisbe', 'quince', 'must', 'pyramus', 'may', 'oberon','’','titania', 'hath', 'enter']
myStopWords = list(punctuation) + stopwords.words('english')+ extrawords
myStopWords
```

```python
wordsNoStop = []
for i in words:
    i = i.lower()  
    if i not in myStopWords:
        wordsNoStop.append(i)
# print(words)
print(wordsNoStop)
```

```python
wordsNoStopWords = [w for w in words if w not in myStopWords]
```

```python
wordsNoStopWords
```

```python
from nltk.probability import FreqDist
```

```python
freq = FreqDist(wordsNoStop)
```

```python
freq
```

```python
#find 20 most frequently used wordxs in the play
for i in sorted(freq, key=freq.get, reverse=True)[:20]:
    print(i,freq[i])
```

```python
#i did my best to remove all the words in the play that wouldn't be found in in the stopword.english since Shakespeare is old english. 
#i also tried to remove as many characters names as possible since they wouldn't give us a sense of subject. 
#Other than that, I think my word list gives a decent sense of what the play is about. 
#With love being the most popular word, we know it's a love story. Words like fairy, eyes, see, night and sweet also give a dreamlike impression of the play which is somewhat accurate to the plot. 
#Otherwise, there are some filler words that don't tell us much. 
```

```python
listOfReviewWords
```

```python
from nltk.stem.porter import PorterStemmer
p_stemmer = PorterStemmer()
```

```python
listOfStemmedWords = []
for i in listOfReviewWords:
    listOfStemmedWords.append([p_stemmer.stem(w) for w in i])
```

```python
listOfStemmedWords[0]
```

```python
!pip install gensim
```

```python
from gensim import corpora, models
import gensim
```

```python
dictionary = corpora.Dictionary(listOfStemmedWords)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in listOfStemmedWords]
```

```python
print(corpus[4])
```

```python
ldamodel = gensim.models.ldamodel.LdaModel(corpus, 
                                           num_topics=2, 
                                           id2word = dictionary, 
                                           passes=20)
```

```python
for i in ldamodel.print_topics(num_topics=3, num_words=20):
    print(i)
```
